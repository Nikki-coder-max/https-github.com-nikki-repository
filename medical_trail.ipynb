```python
from google.colab import files

# Upload the file
uploaded = files.upload()

# Read the uploaded file into a pandas dataframe
import pandas as pd
admin_data = pd.read_csv('Admin.csv')
admin_data.head()  # Show the first few rows
```

```python
labs_data = pd.read_csv('Labs.csv')
labs_data.head()
```

```python
admin_data = pd.merge(admin_data, labs_data, on='ID', how='left')
admin_data.head()
```

```python
# Convert 'Admin Date' to datetime
admin_data['Admin Date'] = pd.to_datetime(admin_data['Admin Date'],errors='coerce')
admin_data
```

```python
# Convert date columns to datetime
admin_data['DRAW_DATE'] = pd.to_datetime(admin_data['DRAW_DATE'], errors='coerce')
admin_data



```

```python
# Convert Units to numeric
admin_data['Units'] = pd.to_numeric(admin_data['Units'], errors='coerce')
admin_data
```

```python
#Convert LAB_VALUE to numeric
admin_data['LAB_VALUE'] = pd.to_numeric(admin_data['LAB_VALUE'], errors='coerce')
admin_data
```

```python
#Monthly Medication Usage
#We will analyze the total number of units administered for each medication each month
# Extract year and month from 'Admin Date'
admin_data['YearMonth'] = admin_data['Admin Date'].dt.to_period('M')
admin_data
```

```python
# Group by 'YearMonth' and 'Med', then sum 'Units'
monthly_usage = admin_data.groupby(['YearMonth', 'Med'])['Units'].sum().reset_index()
monthly_usage
#Explanation: We create a new column that contains the year and month. We then group the data by this new column and the medication type,
#summing the units to find the total usage each month.
```

```python
# Filter data for the relevant months
#Patient Counts on Each Medication
#Next, we'll determine how many patients received each medication from July to November.
monthly_patient_counts = admin_data[admin_data['YearMonth'].dt.month.isin([7, 8, 9, 10, 11])]
patient_counts = monthly_patient_counts.groupby('Med')['ID'].nunique().reset_index()
patient_counts
#Explanation: We filter the data to include only records from July to November, then group by medication type to count the unique patients.
```

```python
# Calculate average monthly dose per patient
#To calculate the average monthly dose per patient for each medication:
# Ensure 'Units' is in string format, replace commas, and then convert to numeric
admin_data['Units'] = pd.to_numeric(admin_data['Units'].astype(str).str.replace(',', ''), errors='coerce')

# Now calculate the average monthly dose
avg_monthly_dose = admin_data.groupby(['YearMonth', 'Med'])['Units'].mean().reset_index()
# Now calculate the average monthly dose
avg_monthly_dose




#We group the data by month and medication type, calculating the average number of units administered
```

```python
# Identify patients who switched medications
switched_patients = admin_data[admin_data['Med'] == 'Medication B']
switched_patients['Switch Month'] = switched_patients['Admin Date'].dt.to_period('M')

switch_counts = switched_patients.groupby('Switch Month')['ID'].nunique().reset_index()
switch_counts
#Explanation: We filter the data for patients who switched to Medication B and create a new column for the month of the switch. We then count the unique patients for each month.
```

```python
# Calculate time on Medication A for switched patients
#Time on Medication A Before Switch
#For patients who switched to Medication B, we calculate how long they were on Medication A.
time_on_med_a = admin_data[admin_data['Med'] == 'Medication A']
time_on_med_a['Duration'] = (time_on_med_a['Admin Date'].max() - time_on_med_a['Admin Date'].min()).days / 7
avg_time_on_med_a = time_on_med_a['Duration'].mean()
avg_time_on_med_a
#Explanation: We calculate the duration that each patient spent on Medication A by finding the difference between the first and last administration dates, converting days to weeks
```

```python
# Average monthly dose before switching to Medication B
avg_dose_a_before = admin_data[admin_data['Med'] == 'Medication A'].groupby('YearMonth')['Units'].mean().mean()

avg_dose_a_before

```

```python
# Average monthly dose after switching to Medication B
avg_dose_b_after = admin_data[admin_data['Med'] == 'Medication B'].groupby('YearMonth')['Units'].mean().mean()
avg_dose_b_after
#Explanation: We compute the average monthly dose for Medication A before switching and Medication B after switching.
```

```python
# Cost of Medication A
cost_per_100_units_a = 1  # $1 for 100 units
cost_per_unit_a = cost_per_100_units_a / 100
cost_per_unit_a

```

```python
# Calculate breakeven price point for Medication B
breakeven_price_b = cost_per_unit_a * (avg_dose_a_before / avg_dose_b_after)
breakeven_price_b
#We calculate the cost per unit for Medication A and then determine the breakeven price for Medication B based on average doses.
```

```python
# Analyze dose change for patients switched in September, October, and November
dose_change = admin_data[(admin_data['Switch Month'].dt.month.isin([9, 10, 11]))]
dose_change_summary = dose_change.groupby(['Switch Month', 'Med'])['Units'].mean().reset_index()
dose_change_summary
```

```python
# Assuming the data is sorted by date, we will calculate the second dose for Medication B
med_b_data = admin_data[admin_data['Med'] == 'Medication B']
med_b_data['Dose Change'] = med_b_data['Units'].diff().fillna(0)  # Calculate change in dose

# Calculate percentage of changes
percentages = med_b_data['Dose Change'].value_counts(normalize=True) * 100
percentages
```

```python
# Compare LAB values for switched patients
lab_comparison = labs_data[labs_data['ID'].isin(switched_patients['ID'])]
average_lab_value_a = lab_comparison[lab_comparison['DRAW_DATE'] < switched_patients['Admin Date'].min()]['LAB_VALUE'].mean()
average_lab_value_b = lab_comparison[lab_comparison['DRAW_DATE'] >= switched_patients['Admin Date'].min()]['LAB_VALUE'].mean()
average_lab_value_b
```
